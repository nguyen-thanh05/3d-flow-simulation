```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
pres_per_step = np.load("pressure_per_timestep.npy")
sat_per_step = np.load("saturation_per_timestep.npy")
prod_rate_per_step = np.load("production_rate_per_timestep.npy")
pres_per_step.shape, sat_per_step.shape, prod_rate_per_step.shape
```

```python
plt.plot(- prod_rate_per_step[:, 1] * 24 * 3600)
plt.grid()
plt.xlabel("Time Step")
plt.ylabel("Oil Production Rate (m3/day)")
plt.title("Oil Production Rate at Producer (10,10,2) Over Time")
```

```python
t = np.array([2, 5, 15, 25])
t = t * 2 - 1
fig, axs = plt.subplots(1, len(t), figsize=(26, 4.5))
for i, ti in enumerate(t):
    im = axs[i].imshow(sat_per_step[ti, :, :, 2], vmin=0, vmax=1, cmap='Blues')
    axs[i].set_title(f'Day {(ti + 1) * 0.5}')
    axs[i].set_xlabel('X')
    axs[i].set_ylabel('Y')
    fig.colorbar(im, ax=axs[i], fraction=0.046, pad=0.04, label='Water Saturation')
```
